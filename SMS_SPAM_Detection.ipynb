```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

nltk.download('punkt')  # Download punkt for tokenization if not already downloaded
nltk.download('stopwords') # Download stopwords if not already downloaded

def transform_text(text):
    """
    Transforms the input text by removing stop words and punctuation.
    """

    stop_words = set(stopwords.words('english'))  # Use a set for efficient lookup
    text = text.lower()  # Convert text to lowercase for case-insensitive matching
    words = word_tokenize(text)  # Tokenize the text into individual words

    # Remove special characters and punctuation marks
    filtered_words = [word for word in words if word.isalnum() and word not in string.punctuation and word not in stop_words]

    return " ".join(filtered_words)  # Join the filtered words back into a string
```

ModelBuilding

we have done data preprocessing

```python
# Naive Bayes classifer works best on textual data
#converting collection of text into a matrix of token counts
cv = CountVectorizer()
X = cv.fit_transform(df['processed']).toarray()
X.shape
```

```python
#storing values of the result column
y=df['result'].values
y
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=49)
```

```python
# creating object for models
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
# training the dataset for guassianNB
gnb.fit(X_train,y_train)
y_pred1 = gnb.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

```python
mnb.fit(X_train,y_train)
y_pred2 = mnb.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
bnb.fit(X_train,y_train)
y_pred3 = bnb.predict(X_test)
print(accuracy_score(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```
